# Preprocessing Cyclical Columns

```python
# Taking cycliality out of flightWeek column, creates to columns

df['flightWeek'] = df['flightWeek'].str[-2:].astype('float')

weeks_year = 52

df['sin_flightWeek'] = np.sin(2*np.pi*df.flightWeek/weeks_year)

df['cos_flightWeek'] = np.cos(2*np.pi*df.flightWeek/weeks_year)

df.drop(columns='flightWeek')
```

```python
# Taking cycliality out of weekOfYear column

df['weekOfYear'] = df['weekOfYear'].str[-2:].astype('float')

weeks_year = 52

df['sin_weekOfYear'] = np.sin(2*np.pi*df.weekOfYear/weeks_year)

df['cos_weekOfYear'] = np.cos(2*np.pi*df.weekOfYear/weeks_year)

df.drop(columns='weekOfYear')
```

```python
airports = pd.read_csv('data/airport_coordinates.csv', delimiter=';')

airports.rename(columns={'Airport Code': 'iata', 'Latitude': 'latitude', 'Longitude': 'longitude'}, inplace=True)

airports['iata'] = airports['iata'].astype('string')
```

```python
# Select only necessary columns
origin_coords = airports[['iata', 'latitude', 'longitude']].copy()
origin_coords.rename(columns={'iata': 'OriginAirport', 'latitude': 'OriginLat', 'longitude': 'OriginLon'}, inplace=True)

destination_coords = airports[['iata', 'latitude', 'longitude']].copy()
destination_coords.rename(columns={'iata': 'DestinationAirport', 'latitude': 'DestLat', 'longitude': 'DestLon'}, inplace=True)

#######

# Merge the origin airport coordinates
df = df.merge(origin_coords, on='OriginAirport', how='left')

# Merge the destination airport coordinates
df = df.merge(destination_coords, on='DestinationAirport', how='left')
```

```python
# Haversine formula

from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    r = 6371  # Radius of Earth in km
    return c * r
```

```python
df['Distance'] = df.apply(lambda row: haversine(row['OriginLat'], row['OriginLon'], row['DestLat'], row['DestLon']), axis=1)

df['Distance'] = df['Distance'].round(1)

df.drop(columns=['OriginLat', 'OriginLon', 'DestLat', 'DestLon'])
```
